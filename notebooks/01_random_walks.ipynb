# Visualizando trayectorias aleatorias
En este notebook vamos a explorar el concepto de **Random Walk** (camino aleatorio), una simulación matemática que describe un camino formado por una sucesión de pasos aleatorios.

## Objetivos

- Visualizar trayectorias utilizando Matplotlib.

```python
# Configuración del entorno
import sys
from pathlib import Path

# Añade la carpeta padre (raíz del proyecto) al path de Python
sys.path.append(str(Path.cwd().parent))

import matplotlib.pyplot as plt

from src.random_walk import RandomWalk
```

### Construcción de la clase RandomWalk
La clase *RandomWalk* encapsula la lógica para generar trayectorias aleatorias. 

La importamos desde el archivo *random_walk.py* en la carpeta *src/* para separar la lógica del análisis. 

### Primera Visualización

```python
# Generamos un random walk.
rw_1 = RandomWalk()
rw_1.trazar_camino()

# Graficamos las posiciones del random walk.
plt.style.use('classic')

fig, ax = plt.subplots()

ax.scatter(rw_1.get_x_values(), rw_1.get_y_values(), s=15)
ax.set_aspect('equal')

plt.show()
```
