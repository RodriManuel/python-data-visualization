# Visualizando trayectorias aleatorias
En este notebook vamos a explorar el concepto de **Random Walk** (camino aleatorio), una simulación matemática que describe un camino formado por una sucesión de pasos aleatorios.

## Objetivos
- Comprender qué es un Random Walk.
- Implementar una simulación bidimensional.
- Visualizar trayectorias utilizando Matplotlib.
- Explorar cómo cambian los resultados entre simulaciones.
- Generar gráficos exportables para el repositorio.

```python
# Configuración del entorno
import sys
from pathlib import Path

# Añade la carpeta padre (raíz del proyecto) al path de Python
sys.path.append(str(Path.cwd().parent))

import matplotlib.pyplot as plt

from src.random_walk import RandomWalk
```

### Construcción de la clase RandomWalk
La clase *RandomWalk* encapsula la lógica para generar trayectorias aleatorias. 

La importamos desde el archivo *random_walk.py* en la carpeta *src/* para separar la lógica del análisis. 

### Primera Visualización

```python
# Generamos un random walk.
rw_1 = RandomWalk()
rw_1.trazar_camino()

# Graficamos las posiciones del random walk.
plt.style.use('classic')

fig, ax = plt.subplots()

ax.scatter(rw_1.get_x_values(), rw_1.get_y_values(), s=15)
ax.set_aspect('equal')

plt.show()
```

### Modificabdo el estilo
Es útil modificar el estilo de este tipo de gráficos para enfatizar las características de cada camino y quitar los elementos que pueden distraernos de lo importante.

Algunas características a enfatizar son:
- En dónde comienza el camino
- En dónde termina el camino
- Hacia dónde se dirigen los pasos

Mientras que vamos a quitar otras como:
- Las etiquetas y números, útiles para otro tipo de gráficos

#### Usamos color map 

```python
# Generamos un random walk.
rw_2 = RandomWalk()
rw_2.trazar_camino()

# Graficamos las posiciones del random walk.
plt.style.use('classic')

fig, ax = plt.subplots()

point_numbers = range(rw_2.number_points)

ax.scatter(rw_2.get_x_values(), rw_2.get_y_values(), c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15)
ax.set_aspect('equal')

plt.show()
```

#### Enfatizamos el inicio y el final

```python
# Generamos un random walk.
rw_3 = RandomWalk()
rw_3.trazar_camino()

# Graficamos las posiciones del random walk.
plt.style.use('classic')

fig, ax = plt.subplots()

point_numbers = range(rw_3.number_points)

ax.scatter(rw_3.get_x_values(), rw_3.get_y_values(), c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15)
ax.set_aspect('equal')

# Enfatizamos el primer y último punto
ax.scatter(0, 0, c='green', edgecolors='none', s=100)
ax.scatter(rw_3.get_x_values()[-1], rw_3.get_y_values()[-1], c='red', edgecolors='none', s=100)

plt.show()
```
