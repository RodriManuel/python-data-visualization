# Generando gráficos con Matplotlib
La visualización de datos permite transformar información numérica en representaciones visuales fáciles de interpretar.

En este notebook voy a generar mis primeros gráficos utilizando Matplotlib.

## Objetivo
- Crear gráficos básicos con Matplotlib
- Personalizar estilos visuales
- Generar datos automáticamente
- Explorar scatter plots
- Utilizar colormaps
- Exportar visualizaciones

```python
from pathlib import Path
import matplotlib.pyplot as plt
```

## Primer gráfico líneal
Comenzamos visualizando la sucesión de los cuadrados de los números naturales del 1 al 5 para comprender cómo Matplotlib representa relaciones entre variables.

```python
cuadrados = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()

ax.plot(cuadrados)

plt.show()
```

El gráfico muestra cómo el crecimiento cuádratico aumenta progresivamente a medida que avanzan los valores de entrada.

Aunque el gráfico nos muestra que los números van incrementandose, el estilos visual es bastante pobre:
- Los números de los ejes X e Y son muy pequeños.
- La línea del gráfico es demasiado fina.
- No existen títulos que permitan al gráfico explicarse por si mismo.
- Además el eje X arranca desde 0.

Por suerte, Matplotlib nos permite ajustar cualquier aspecto de un gráfico. Esto vamos a estar haciendo en las siguientes secciones.

### Mejorando el estilo visual
Vamos a mejorar el estilo de los siguientes elementos en un gráfico:
- Grosor de la línea.
- Título para el gráfico.
- Subtítulos para los ejes X e Y.
- Tamaño de fuente.

```python
valores_eje_y = [1, 2, 3, 4, 5] # Si no lo agregamos, matplotlib establece valores arrancando en 0.
cuadrados = [1, 4, 9, 16, 25]

plt.style.use('seaborn-v0_8') # Estilo del fondo

fig, ax = plt.subplots(figsize=(10, 6)) # Establece el ancho y alto del gráficos.
ax.plot(valores_eje_y, cuadrados, linewidth=3) # El segundo parámetro determina el grosor de la línea.

# Establecemos títulos y subtítulos.
ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

# Establecemos el tamaño de los números en los ejes X e Y.
ax.tick_params(labelsize=14)

plt.show()
```

## Gráficos de puntos con scatter()
Hasta ahora venimos generando gráficos de línea utilizando la función plot().

Aveces puede ser útil gráficar cada posición del gráfico usando puntos individuales. Por ejemplo, podríamos gráficar los valores más bajos en un color y los más altos en otro.

La forma de graficar un solo punto es pasandole a scatter() las coordenadas de los ejes X e Y:

```python
plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots()
ax.scatter(2, 4, s=200)

# Establece títulos y subtítulos.
ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

# Establece el tamaño de los números en los ejes X e Y.
ax.tick_params(labelsize=14)

plt.show()
```

### Graficando una serie de puntos
Al igual que con los gráficos de línea, le pasamos a scatter() dos listas separadas para los valores X e Y.

```python
x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 24]

plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x_values, y_values, s=100)

# Establece títulos y subtítulos.
ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

# Establece el tamaño de los números en los ejes X e Y.
ax.tick_params(labelsize=14)

plt.show()
```

## Generar datos automáticamente
Escribir las listas a mano es tedioso e ineficiente. Especialmente si debemos graficar una gran cantidad de puntos.
Podemos automatizar este proceso usando un for y listas comprimidas para generar los valores para los ejes X e Y.

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x_values, y_values, s=10)

# Establecemos títulos y subtítulos.
ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

# Establecemos el tamaño de los números en los ejes X e Y.
ax.tick_params(labelsize=14)

#Establecemos el rango para cada eje.
ax.axis([0, 1100, 0, 1_100_000])

plt.show()
```

### Mejorando el estilo del gráfico anterior
- Tick labels
- Custom colors
- Color map

```python
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)

# Establecemos títulos y subtítulos.
ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

# Establecemos el tamaño de los números en los ejes X e Y.
ax.tick_params(labelsize=14)

#Establecemos el rango para cada eje.
ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style='plain')

plt.show()
```

## Exportación de gráficos
Para descargar los gráficos y guardarlos en la carpeta adecuada automáticamente, vamos a usar
- el módulo pathlib de Python 
- y la función savefig() de Matplotlib

Ahora vamos a descargar 4 de los gráficos que venimos generando:

### Empezamos con los dos primeros gráficos líneales

```python
cuadrados = [1, 4, 9, 16, 25]

# Cramos la ruta hacia la carpeta destino y la creamos si no existe
output_dir = Path("../outputs/images")
output_dir.mkdir(parents=True, exist_ok=True)

fig, ax = plt.subplots()

ax.plot(cuadrados)

# Guardamos el gráfico y definimos el nombre de la imágen
output_file = output_dir / "line_plot_1.png"

plt.savefig(output_file, bbox_inches="tight") # Usamos la función savefig() en lugar de show()
```

```python
valores_eje_y = [1, 2, 3, 4, 5]
cuadrados = [1, 4, 9, 16, 25]

# Cramos la ruta hacia la carpeta destino y la creamos si no existe
output_dir = Path("../outputs/images")
output_dir.mkdir(parents=True, exist_ok=True)

plt.style.use('seaborn-v0_8') 

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(valores_eje_y, cuadrados, linewidth=3)

ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

ax.tick_params(labelsize=14)

# Guardamos el gráfico y definimos el nombre de la imágen
output_file = output_dir / "line_plot_2.png"

plt.savefig(output_file, bbox_inches="tight") # Usamos la función savefig() en lugar de show()
```
