# Generando gráficos con Matplotlib
La visualización de datos permite transformar información numérica en representaciones visuales fáciles de interpretar.

En este notebook voy a generar mis primeros gráficos utilizando Matplotlib.

## Objetivo
- Crear gráficos básicos con Matplotlib
- Personalizar estilos visuales
- Generar datos automáticamente
- Explorar scatter plots
- Utilizar colormaps
- Exportar visualizaciones

```python
from pathlib import Path
import matplotlib.pyplot as plt
```

## Primer gráfico líneal
Comenzamos visualizando la sucesión de los cuadrados de los números naturales del 1 al 5 para comprender cómo Matplotlib representa relaciones entre variables.

```python
cuadrados = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()

ax.plot(cuadrados)

plt.show()
```

El gráfico muestra cómo el crecimiento cuádratico aumenta progresivamente a medida que avanzan los valores de entrada.

Auneque el gráfico nos muestra que los números van incrementandose, el estilos visual es bastante pobre:
- Los números de los ejes X e Y son muy pequeños.
- La línea del gráfico es demasiado fina.
- No existen títulos que permitan al gráfico explicarse por si mismo.
- Además el eje X arranca desde 0.

Por suerte, Matplotlib nos permite ajustar cualquier aspecto de un gráfico. Esto vamos a estar haciendo en las siguientes secciones.

## Mejorando el estilo visual
Vamos a mejorar el estilo de los siguientes elementos en un gráfico:
- Grosor de la línea.
- Título para el gráfico.
- Subtítulos para los ejes X e Y.
- Tamaño de fuente.

```python
valores_eje_y = [1, 2, 3, 4, 5] # Si no lo agregamos, matplotlib establece valores arrancando en 0.
cuadrados = [1, 4, 9, 16, 25]

plt.style.use('Solarize_Light2') # Estilo del fondo

fig, ax = plt.subplots(figsize=(10, 6)) # Establece el ancho y alto del gráficos.
ax.plot(valores_eje_y, cuadrados, linewidth=3) # El segundo parámetro determina el grosor de la línea.

# Establece títulos y subtítulos.
ax.set_title("Números cuadrados", fontsize=24)
ax.set_xlabel("Valores", fontsize=14)
ax.set_ylabel("Cuadrados", fontsize=14)

# Establece el tamaño de los números en los ejes X e Y.
ax.tick_params(labelsize=14)

plt.show()
```
